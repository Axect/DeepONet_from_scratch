```python
# Basic
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# IO
import netCDF4 as nc
```

## Read data file

```python
data = nc.Dataset("data/grf_150_101_0.30.nc").variables
```

```python
keys = data.keys()
values = data.values()
```

```python
data_dict = {}

for k, v in zip(keys, values):
    data_dict[k] = np.array(v)
```

```python
df = pd.DataFrame(data_dict)
```

```python
df
```

```python
plt.figure(figsize=(10, 6), dpi=150)
for i in range(5):
    plt.plot(df["x"], df[f"y{i}"])
plt.show()
```
